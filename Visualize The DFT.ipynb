I was playing a bit around with the DFT, and I discovered (rather <a href="">rediscovered</a>) several ways beautiful visualizations of the DFT. Besides being rather cool it might provide some intuitive insight into the transform: 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import dft

fig, ax = plt.subplots(3, 8, figsize=(13, 6))

ax[0,0].set_ylabel("Real")
ax[1,0].set_ylabel("Imag")
ax[2,0].set_ylabel("Abs")

for i in range(0, 8):
    n = 2**(i+2)
    F_n = dft(n)
    ax[0,i].imshow(np.real(F_n))
    ax[1,i].imshow(np.imag(F_n))
    ax[2,i].imshow(np.abs(F_n))
    ax[0,i].set_title("n=%i"%n)
    
    for j in range(3): 
        ax[j,i].set_yticklabels([])
        ax[j,i].set_xticklabels([])
        ax[j,i].set_yticks([])
        ax[j,i].set_xticks([])

plt.tight_layout()
```
